# Polynomial Regression


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(42)
n_samples = 100

X = np.linspace(0, 10, 100)
rng = np.random.rand(n_samples) * 100

y = X ** 3 + rng + 100

plt.figure(figsize=(10, 10))
plt.scatter(X, y)
plt.show()
```

# Linear Regression 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
lr.fit(X.reshape(-1, 1), y)
model_pred = lr.predict(X.reshape(-1, 1))

plt.figure(figsize=(16, 12))
plt.scatter(X, y)
plt.plot(X, model_pred)
print(r2_score(y, model_pred))
plt.show()
```

# Polynomial Regression 

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=3)
X_poly = poly_reg.fit_transform(X.reshape(-1,1))
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
y_pred = lin_reg_2.predict(X_poly)
```

```python
plt.figure(figsize=(12, 10))
plt.scatter(X, y)
plt.plot(X, y_pred)
print(r2_score(y, y_pred))
plt.show()
```

# Boston Housing Data 

```python
df_boston = pd.read_csv('housing.data', delim_whitespace=True, header=None)
df_boston.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
```

```python
df_boston.corr()
```

```python
X_boston = df_boston['DIS'].values
y_boston = df_boston['NOX'].values
```

```python
plt.figure(figsize=(12, 10))
plt.scatter(X_boston, y_boston)
plt.show();
```

# Linear Regression 

```python
lr_boston = LinearRegression()
lr_boston.fit(X_boston.reshape(-1,1), y_boston)
model_pred = lr_boston.predict(X_boston.reshape(-1,1))
```

```python
plt.figure(figsize=(12, 10))
plt.scatter(X_boston, y_boston)
plt.plot(X_boston, model_pred)
print(r2_score(y_boston, model_pred))
plt.show()
```

# Quadratic Regression 

```python
quad_reg = PolynomialFeatures(degree=2)
X_quad = quad_reg.fit_transform(X_boston.reshape(-1, 1))
```

```python
quad_lr = LinearRegression()
quad_lr.fit(X_quad, y_boston)

X_fit = np.arange(X_boston.min(), X_boston.max(), 1)[:, np.newaxis]
quad_model_pred = quad_lr.predict(quad_reg.fit_transform(X_fit.reshape(-1, 1)))


plt.figure(figsize=(12, 10))
plt.scatter(X_boston, y_boston)
plt.plot(X_fit, quad_model_pred)
print(r2_score(y_boston, quad_lr.predict(X_quad)))
plt.show()
```

# Cubic Regression 

```python
cube_reg = PolynomialFeatures(degree=3)
X_cube = cube_reg.fit_transform(X_boston.reshape(-1, 1))
```

```python
cube_lr = LinearRegression()
cube_lr.fit(X_cube, y_boston)

X_fit = np.arange(X_boston.min(), X_boston.max(), 1)[:, np.newaxis]
cube_model_pred = cube_lr.predict(cube_reg.fit_transform(X_fit.reshape(-1, 1)))


plt.figure(figsize=(12, 10))
plt.scatter(X_boston, y_boston)
plt.plot(X_fit, cube_model_pred)
print(r2_score(y_boston, cube_lr.predict(X_quad)))
plt.show()
```
